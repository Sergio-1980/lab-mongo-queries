### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
# Your Code
select_02 = {"number_of_employees": {"$gt": 5000}}

more_5000 = list(c.find(select_02).limit(20).sort("number_of_employees",1))
```

```python
more_5000
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

```python
# Your Code
select_03 = {"founded_year": {"$gt": 2000}}
select_04 = {"founded_year": {"$lt": 2005}}
retrieve = {"_id": 0, "name": 1,  "founded_year": 1}

between_00_05 = list(c.find({"$and": [select_03, select_04]}, retrieve))
```

```python
between_00_05
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
# Your Code
select_05 = {"ipo.valuation_amount": {"$gt": 100000000}}
select_06 = {"founded_year": {"$lt": 2010}}
retrieve = {"_id": 0, "name": 1,  "ipo": 1}

valuation_amount = list(c.find({"$and": [select_05, select_06]}, retrieve))
```

```python
valuation_amount
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
# Your Code
select_07 = {"number_of_employees": {"$lt": 1000}}
select_08 = {"founded_year": {"$lt": 2005}}

companies_count = list(c.find({"$and": [select_07, select_08]}).sort("number_of_employees",1).limit(10))
```

```python
companies_count
```
